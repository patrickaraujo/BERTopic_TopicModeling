```python
data = pd.read_csv('Datasets/Duolingo_en.csv')
#data = data[:1000]
```

```python
data = data[data['content'].str.len() > 30].reset_index()
```

```python
data.head(10)
```

```python
data
```

```python
reviews = data['content'].values
```

```python
reviews
```

```python
import nltk
from gensim.models import Word2Vec, KeyedVectors
```

```python
reviewsVec = [nltk.word_tokenize(comm) for comm in reviews]
model = Word2Vec(reviewsVec, min_count=1, vector_size=32)
```

```python
model
```

```python
# Get the vector for a specific word
vector = model.wv['word']
print(vector)
```

```python
# Find similar words
similar_words = model.wv.most_similar('word')
print(similar_words)
```

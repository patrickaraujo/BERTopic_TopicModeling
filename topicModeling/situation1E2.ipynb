```python
merged_df = pd.concat([data0, data1, data2])
merged_df = merged_df.reset_index(drop=True)
```

```python
merged_df
```

```python
newData = pd.DataFrame(columns=['topic', 'sentiment'])
```

```python
import ast
if all:
    for i in merged_df.index:
        listString = merged_df['topics'][i]
        words_list = ast.literal_eval(listString)
        for word in words_list:
            if word != "":
                new_row = {'topic': word, 'sentiment': 0}
                newData = pd.concat([newData, pd.DataFrame([new_row])], ignore_index=True)
                new_row = {'topic': word, 'sentiment': 1}
                newData = pd.concat([newData, pd.DataFrame([new_row])], ignore_index=True)
                new_row = {'topic': word, 'sentiment': 2}
                newData = pd.concat([newData, pd.DataFrame([new_row])], ignore_index=True)
else:
    for i in merged_df.index:
        listString = merged_df['topics'][i]
        words_list = ast.literal_eval(listString)
        for word in words_list:
            if word != "":
                new_row = {'topic': word, 'sentiment': merged_df['sentiment'][i]}
                newData = pd.concat([newData, pd.DataFrame([new_row])], ignore_index=True)
```

```python
newData
```

```python
newData = newData.sample(frac=1).reset_index(drop=True)
```
